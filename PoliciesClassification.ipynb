```python
full_df = pd.read_csv("/content/Policy.csv")
df = full_df[["Policy"]]
df["Policy"] = df["Policy"].astype(str)
full_df.head()
```

```python
full_df[["PolicyType"]].count()
```

```python
# Lower Case
df["policy_lower"] = df["Policy"].str.lower()
df.head()
```

```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["policy_wo_punct"] = df["policy_lower"].apply(lambda text: remove_punctuation(text))
df.head()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["policy_wo_stop"] = df["policy_wo_punct"].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
from collections import Counter
cnt = Counter()
for text in df["policy_wo_stop"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df["policy_wo_stopfreq"] = df["policy_wo_stop"].apply(lambda text: remove_freqwords(text))
df.head()
```

```python
from nltk.stem.porter import PorterStemmer

# Drop the two columns 
# df.drop(["policy_wo_stopfreq"], axis=1, inplace=True) 

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["policy_stemmed"] = df["policy_wo_stopfreq"].apply(lambda text: stem_words(text))
df.head()
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
lemmatizer = WordNetLemmatizer()
```

```python
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

df["policy_lemmatized"] = df["policy_stemmed"].apply(lambda text: lemmatize_words(text))
df.head()
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
df['tokenized'] = df['policy_lemmatized'].apply(word_tokenize)
df.head()
```

```python
punc = string.punctuation
df['policy_no_punc'] = df["tokenized"].apply(lambda x: [word for word in x if word not in punc])
df.head()
```

```python
df['policy_preprocess_data'] = [' '.join(map(str, l)) for l in df['policy_no_punc']]
df.drop(["tokenized"], axis=1, inplace=True) 
df.drop(["policy_lemmatized"], axis=1, inplace=True) 
df.drop(["policy_stemmed"], axis=1, inplace=True) 
df.drop(["policy_wo_stopfreq"], axis=1, inplace=True) 
df.drop(["policy_wo_stop"], axis=1, inplace=True) 
df.drop(["policy_wo_punct"], axis=1, inplace=True) 
df.drop(["policy_lower"], axis=1, inplace=True) 
df.head()
```

```python
full_df["Policy"] = df['policy_preprocess_data']
full_df["Policy_tokenized"] = df['policy_no_punc']
```
